# Functional programming
Functional programming is an advaced programming topic, so we only introduce some basic ideas:
- A function is like other data types (int, float, etc). Can be stored in a variable, passed as parameter, etc.
- You can pass a function as parameter to modify the behavior

Example, sorting a list.

```python
stone_names = ["Quartz", "Amethyst", "Diamond", "Ruby", "Emerald", "Sapphire", "Topaz", "Opal", "Jade", "Citrine"]
sorted(stone_names)
```

What if I want to sort by other characteristic, like the last letter?
- sorted receives a parameter that is a function.
    - This function extract something for each list value
    - The sort is performed over that value

```python
def to_sort(v):
    return v[-1]

sorted(stone_names, key=to_sort)
```

```python
# Sort by the number of characters
def to_sort(v):
    return len(v)

sorted(stone_names, key=to_sort)
```

```python
# Sort by the number of characters, reverse
sorted(stone_names, key=to_sort, reverse=True)
```

The **sorted** method receives a function parameter that allows to modify the **sorted** behavior

```python
# Other example
def to_sort(v):
    return v[0]

students = [("Alice", 88), ("Bob", 92), ("Charlie", 85)]
sorted(students, key=to_sort)
```

The use of a function to only pass as parameter can be improved by anonymous functions

```python
sorted(students, key=lambda x:x[0])
```

```python
sorted(students, key=lambda x:x[1])
```

## Other examples

Min/Max of a sequence

```python
students = [("Alice", 88), ("Bob", 92), ("Charlie", 85)]

# student with maximum value
max(students, key=lambda x:x[1])
```

```python
max(students, key=lambda x:x[0])
```

```python
max(students, key=lambda x:x[0][1])
```

```python
min(students, key=lambda x:x[0][1])
```

Mapping an action

```python
stone_names = ["Quartz", "Amethyst", "Diamond", "Ruby", "Emerald", "Sapphire", "Topaz", "Opal", "Jade", "Citrine"]
list(map(lambda x: x[0], stone_names))
```
