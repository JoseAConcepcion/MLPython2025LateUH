# Dictionaries
A dictionary is a data structure that stores key-value pairs
- values can be of any type
- keys can be any inmutable type
- keys cannot be repited, values can

```python
my_dict = {'name': 'John', 'age': 28, 'profession': 'programmer'}
my_dict
```

You can retrieve a value associated to a key very fast, no matter how long the dictionary is

```python
my_dict['name']
```

```python
my_dict['age']
```

Access a dictionary using a missing key generates an exception. Solutions:
- Test if the element is there using **in**
- Use get method

```python
'name' in my_dict
```

```python
result = None
if 'name' in my_dict:
    result = my_dict['name']
print(result)
```

```python
result = None
if 'namex' in my_dict:
    result = my_dict['name']
print(result)
```

```python
# Using safe method get
result = my_dict.get('name')
print(result)
```

```python
result = my_dict.get('namex')
print(result)
```

```python
# You can control the value to assing if key is missing
result = my_dict.get('namex', 'no name')
print(result)
```

We can add pairs to a dictionary or modify existing pairs by the key.

```python
my_dict
```

```python
my_dict['location'] = 'Berlin'
my_dict
```

```python
# If key exist, value is updated
my_dict['location'] = 'Holzminden'
my_dict
```

```python
# Remove an element from a dictionary have a funny syntax
del my_dict['age']
my_dict
```

## Iterating a dictionary by **for** loop

```python
for k in my_dict:
    print(k)
```

```python
for v in my_dict.values():
    print(v)
```

```python
for k, v in my_dict.items():
    print(k, v)
```
