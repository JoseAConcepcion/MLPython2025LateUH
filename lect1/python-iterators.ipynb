# Iterators
In Python, iterators allow to travel all the elements of a sequence in order.
- It is the mechanism behind the powerfull **for** instruction

```python
for i in range(10):
    print(i)
```

```python
# range do not build a sequence, but provide it one by one
range(10)
```

For iterating an iterator you need a for instruction, or a list constructor

# Useful iterators
## range

```python
# range(end)
for i in range(8):
    print(i)
```

```python
# range(ini, end)
for i in range(3, 10):
    print(i)
```

```python
# range(ini, end, step)
for i in range(3, 10, 2):
    print(i)
```

```python
# Todos los resultados del iterador a la lista
l = list(range(3, 10, 2))
print(type(l))
l
```

## enumerate
Supose you want to select the elements in a list in odd positions
- You cannot use directly comprehensions. Why?

```python
values = [3, 4, 5, 7, 1, 2, 9, 8]
result = []
for v in range(len(values)):
    if v % 2 == 1:
        result.append(values[v])
result
```

```python
# solution, enumerate
for idx, v in enumerate(values):
    print(idx, v)
```

```python
[v for idx, v in enumerate(values) if idx%2 == 1]
```

## zip
Supose you have 'parallel lists', containing information in the same index related. You want to process them together.

```python
# given the lists, return the names of persons older than 20 years old
names = ['peter', 'mary', 'john', 'claude']
ages = [23, 34, 25, 16]

result = []
for i in range(len(names)):
    if ages[i] > 20:
        result.append(names[i])

result
```

```python
# zip, enumerate parallel lists
for name, age in zip(names, ages):
    print(name, age)
```

```python
[name for name, age in zip(names, ages) if age>20]
```

```python
# zip can be used with any number of lists
colors = ['pink', 'white', 'red', 'black']
for name, age, color in zip(names, ages, colors):
    print(name, age, color)
```

```python
# zip stops when one of the sequences ends
sizes = [123, 134, 115]
for name, size in zip(names, sizes):
    print(name, size)
```

## reversed

```python
colors = ['pink', 'white', 'red', 'black']
for c in reversed(colors):
    print(c)
```

## sorted

```python
values = [3, 4, 5, 7, 1, 2, 9, 8]
for v in sorted(values):
    print(v)
```
